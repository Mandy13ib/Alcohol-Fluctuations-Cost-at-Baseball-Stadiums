```python
# Python SQL toolkit and Object Relational Mapper
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python
engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5433/MLB")
conn = engine.connect()
```

```python
# Query All Records in the the Database
beer = pd.read_sql("SELECT * FROM mlb_beer_prices", conn)
beer
```

```python
wins = pd.read_sql("SELECT * FROM mlb_wins", conn)
wins
```

```python
mlb_beer_vs_wins = pd.merge(beer, wins, on = 'team')
mlb_beer_vs_wins
```

```python
del mlb_beer_vs_wins['nickname']
```

```python
del mlb_beer_vs_wins['year_y']
```

```python
 mlb_beer_vs_wins
```

```python
mlb_beer_vs_wins = mlb_beer_vs_wins.rename(columns = {"year_x": 'Year',
                                       "team": 'Team',
                                       "city": 'City',
                                       "price" : 'Price of Beer',
                                       "size": 'Size of Beer',
                                       "price_per_ounce": 'Price per Ounce',
                                       "number_of_games" : 'Number of Games Played',
                                       "win": 'Number of Games Won'})
```

```python
mlb_beer_vs_wins
```

```python
beer_and_wins = pd.read_sql("SELECT * FROM mlb_combined_data", conn)
beer_and_wins
```

```python
beer_and_wins = beer_and_wins.rename(columns = {"year_x": 'Year',
                                       "team": 'Team',
                                       "city": 'City',
                                       "price" : 'Price of Beer',
                                       "size": 'Size of Beer',
                                       "price_per_ounce": 'Price per Ounce',
                                       "number_of_games" : 'Number of Games Played',
                                       "wins": 'Number of Games Won'})
beer_and_wins
```
